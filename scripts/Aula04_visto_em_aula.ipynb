```python
#criando matriz numpy
import numpy as np
a=np.zeros((5,5))
print(a)
print('Indexaçao numpy')
print(a[0,0])
```

```python
#criando número aleatórios
b=np.random.rand(5,5)
print(b)
#shape dimnesão da matriz
print(b.shape)
print(b)
Vmin=1
Vmax=100
c=Vmin+(Vmax-Vmin)*b
print(c)
```

```python
#plotar matriz
import matplotlib.pyplot as plt
plt.plot(c)
#inverter a matriz
plt.plot(c.transpose())
#plotar uma linha
plt.plot(c[0,:])
```

```python
#size irá mostrar quantas linhas tem na coluna 0
x = range(0,np.size(c[0,:]))
print(x)
x=np.array(range(0,np.size(c[0,:])))
plt.title('Gráfico com valores inteiros no eixo x')
#Definindo os valores
plt.xticks(x)
```

```python
#criando um gráfico de barras
fig, ax=plt.subplots(3,1)

# Plotando os gráficos de barras nos subplots
ax[0].bar(x, c[:, 0])  # Usando todos os valores da primeira coluna de c
ax[1].bar(x, c[:, 1])  # Usando todos os valores da segunda coluna de c
ax[2].bar(x, c[:, 2])  # Usando todos os valores da terceira coluna de c

# Adicionando títulos aos subplots
ax[0].set_title('Primeira Coluna de c')
ax[1].set_title('Segunda Coluna de c')
ax[2].set_title('Terceira Coluna de c')

# Ajustando o layout
#plt.tight_layout() -> ajusta os subplots para ficar melhor dentro da figura

plt.tight_layout()
```

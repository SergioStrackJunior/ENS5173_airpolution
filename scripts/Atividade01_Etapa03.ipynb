Objetivo: Produção de bruto aço no estado de São Paulo ao longo de 5 anos.

```python
import pandas as pd
import numpy as np

# Dados fornecidos de 3 anos
anos = [2018, 2019, 2020, 2021, 2022]
valores = [2.382, 2.265, 2.613, 2.831, 2.691]

# Gerando números aleatórios para os anos de 2021 a 2050
anosfuturos = list(range(2023, 2050))
valoresfuturos = [np.random.uniform(min(valores), max(valores)) for _ in range(len(anosfuturos))]

dados = pd.DataFrame({'Ano': anos + anosfuturos, 'Valor': valores + valoresfuturos})

# Exibindo a tabela
print(dados)
```

```python
import matplotlib.pyplot as plt

# Criando o gráfico
plt.plot(anos, valores, marker='o', linestyle='-')

# Adicionando rótulos aos eixos
plt.xlabel('Ano')
plt.ylabel('Valor')

# Adicionando título ao gráfico
plt.title('Produção de aço bruto')

# Exibindo o gráfico
plt.grid(True)
plt.show()

#Isso renderiza o gráfico na tela para visualização.
```

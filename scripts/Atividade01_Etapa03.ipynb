Objetivo: Gerar números aleatórios que representem a atividade industrial (ex: toneladas de cimento produzido). Utilize alguma indústria do estado para basear seus códigos.
Utilizado dados de emplacamentos de veículos leves no Brasil ao longo de 9 anos.

```python
import pandas as pd
import numpy as np

# Dados fornecidos
anos = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
valores = [2463414, 1986502, 2170298, 2468068, 2658167, 1949892, 1974840, 1952794, 2176008]

# Gerando números aleatórios para os anos de 2024 a 2050
anos_futuros = list(range(2024, 2050))
valores_futuros = [np.random.uniform(min(valores), max(valores)) for _ in range(len(anos_futuros))]

dados = pd.DataFrame({'Ano': anos + anos_futuros, 'Valor': valores + valores_futuros})

# Exibindo a tabela
print(dados)
```

```python
import matplotlib.pyplot as plt

# Criando o gráfico
plt.plot(anos, valores, marker='o', linestyle='-')

# Adicionando rótulos aos eixos
plt.xlabel('Ano')
plt.ylabel('Valor')

# Adicionando título ao gráfico
plt.title('Emplacamento ao longo dos anos')

# Exibindo o gráfico
plt.grid(True)
plt.show()

#Isso renderiza o gráfico na tela para visualização.
```
